```python
!pip install opencv-contrib-python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= './opencv_logo.png'
y= './3-1_dVenus_0250.tif'
```

```python
opencv= cv2.imread(x)
plt.imshow(opencv)
```

# smoothing out the cell regions

```python
brain_slice=cv2.imread(y)
plt.imshow(brain_slice)
```

```python
blur = cv2.blur(brain_slice,(10,10))
```

```python
plt.subplot(121),plt.imshow(brain_slice),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cv2.imwrite('blured.tif',blur)
```

# subtract cell region - blurred

```python
dif =cv2.subtract(brain_slice,blur)
cv2.imwrite('diff.tif',dif)
```

```python
plt.imshow(dif)
```

# average out the regions of interest

## still the region is not clear , we fic it to 5X5 pixel , dim the centre and effectively create a circle

```python
kernel = np.ones((5,5),np.float32)
kernel[2,2]=0
kernel=kernel/24
kernel
```

```python
dst = cv2.filter2D(brain_slice,-1,kernel)
cv2.imwrite('avg.tif',dst)
dif2=cv2.subtract(brain_slice,dst)
cv2.imwrite('dif2.tif',dif2)
```

# testing the various gradients in opencv

```python
laplacian = cv2.Laplacian(brain_slice,cv2.CV_64F)
sobelx = cv2.Sobel(brain_slice,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(brain_slice,cv2.CV_64F,0,1,ksize=5)

plt.subplot(2,2,1),plt.imshow(brain_slice,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

# apply gradients smooth to the differce

```python
laplacian = cv2.Laplacian(dif2,cv2.CV_64F)
sobelx = cv2.Sobel(dif2,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(dif2,cv2.CV_64F,0,1,ksize=5)

plt.subplot(2,2,1),plt.imshow(dif2,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
cv2.imwrite('laplacian.png',laplacian)
cv2.imwrite('sobelx.png',sobelx)
cv2.imwrite('sobely.png',sobely)
```

# coloring the cell boundary and superimpose / adding the  matrxi

```python
laplacian=cv2.Laplacian(dif,cv2.CV_32F)


r= cv2.cvtColor(laplacian,cv2.COLOR_BGR2GRAY)
g= np.zeros((2160,2560),np.float32)
b= np.zeros((2160,2560),np.float32)


cell_egde=cv2.merge((g,b,r))
```

```python
overlay= brain_slice+cell_egde
cv2.imwrite('overlay.png',overlay)
```
